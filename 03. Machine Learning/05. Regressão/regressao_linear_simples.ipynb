```python
x = df.Salario.values.reshape(-1, 1)
y = df.Limite_Emprestimo.values.reshape(-1, 1)

from sklearn.model_selection import train_test_split
x_treinamento,x_teste ,y_treinamento,y_test = train_test_split(
    x
    ,y
    ,test_size=0.3
)
x_treinamento
```

```python
from sklearn.linear_model import LinearRegression

funcao = LinearRegression()
funcao.fit(x_treinamento, y_treinamento)
funcao.score(x_treinamento, y_treinamento)  # Avalia no mesmo conjunto
```

```python
plt.scatter(x_treinamento, y_treinamento, label="Treinamento", color="blue")
plt.scatter(x_teste, y_test, label="Teste", color="green")
plt.plot(x_teste, funcao.predict(x_teste), color='red', label="RegressÃ£o")

plt.legend()
plt.title("GrÃ¡fico da RegressÃ£o Linear")
plt.xlabel("SalÃ¡rio")
plt.ylabel("Limite de EmprÃ©stimo")
plt.grid(True)
plt.show()
```
