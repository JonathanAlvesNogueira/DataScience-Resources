```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"D:\donwloads\housing.csv",
                    sep=',', encoding='iso-8859-1')

#  Número de linhas e colunas
display('Número de linhas e colunas')
display(df.shape)

df.rename(columns={
    'RM': 'media_comodos',
    'LSTAT': 'percentual_classe_baixa',
    'PTRATIO': 'razao_aluno_professor',
    'MEDV': 'valor_medio_casas'
}, inplace=True)

display(df)
```

```python
# Invés de fazer todas as regressão existente, vamos juntas as 3 primeiras colunas e montar uma regressão delas

independente  = df.iloc[:, 0:3].values
independente

dependente = df.iloc[:, 3].values
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(independente, dependente, test_size=0.3, random_state=0)
```

```python
multipla = LinearRegression()
multipla.fit(x_treino, y_treino)
```

```python
multipla.intercept_
multipla.coef_
```

```python
multipla.score(x_treino, y_treino)
```

```python
previsao = multipla.predict(x_teste)
previsao
```

```python
multipla.score(x_teste, y_teste)
```

```python
valor_casa = multipla.predict([[8, 25, 15]])
valor_casa
```
