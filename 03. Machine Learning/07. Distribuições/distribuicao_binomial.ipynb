```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Configurações
n = 50    # transações por lote
p = 0.02  # taxa de fraude esperada (2%)

# Gerando todos os cenários possíveis (0 a 10 fraudes para o gráfico ficar legível)
x = np.arange(0, 11)
probabilidades = binom.pmf(x, n, p)

# Visualização
plt.bar(x, probabilidades, color='skyblue', edgecolor='black')
plt.title(f'Distribuição de Fraudes (n=50, p=2%)')
plt.xlabel('Número de Fraudes encontradas no lote')
plt.ylabel('Probabilidade')
plt.xticks(x)
plt.show()

# A Prova Real: Qual a chance de soar o alarme (5 ou mais)?
# Usamos 1 - cdf(4), que significa: 100% menos a chance de dar 0, 1, 2, 3 ou 4.
chance_alarme = 1 - binom.cdf(4, n, p)
print(f"Chance do alarme tocar por 'azar' (falso positivo): {chance_alarme:.4%}")
```
