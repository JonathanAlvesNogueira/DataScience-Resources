```python
lista_links = df['link'].to_list()
lista_links
```

```python
import time
import random
import undetected_chromedriver as uc

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = uc.Chrome(version_main=148)
wait = WebDriverWait(driver, 15)

try:
    for url in lista_links:
        dados = []
        print(f"Acessando {url}...")
        driver.get(url)

        wait.until(
            EC.presence_of_all_elements_located((By.TAG_NAME, "p"))
        )

        time.sleep(random.uniform(3, 6))
        elementos = driver.find_elements(By.TAG_NAME, "p")

        for el in elementos:
            texto = el.text.strip()
            if texto and len(texto) > 10:
                dados.append({
                    "texto": texto
                })

        # display(dados)
        palavras_lixo = [
            "compartilhe",
            "está com problema",
            "qual a posição",
            "lista de melhores",
            "confira a classificação",
            "área restrita",
            "acesse aqui",
            "conteúdos destinados",
            "classificação indicativa"
        ]

        registro = {
            "link": url,
            "ds_situacao_reclamacao": None,
            "estado_reclamacao": None,
            "data_reclamacao": None,
            "id_reclamacao": None,
            "descricao_reclamacao": None
        }

        for d in dados:
            texto = d["texto"].lower()

            # 🔥 remove lixo
            if any(p in texto for p in palavras_lixo):
                continue

            # status
            if "respondida" in texto:
                registro["ds_situacao_reclamacao"] = texto

            # estado (curto)
            elif " - " in texto and len(texto) < 50:
                registro["estado_reclamacao"] = texto

            # data
            elif "/" in texto and "às" in texto:
                registro["data_reclamacao"] = texto

            # id
            elif "id:" in texto:
                registro["id_reclamacao"] = texto.replace("id: ", "")

            # descrição (texto grande real)
            elif len(texto) > 100:
                registro["descricao_reclamacao"] = texto
            
            # descrição (texto grande real respondido pela empresa )
            elif len(texto) > 100:
                registro["descricao_reclamacao"] = texto
            
        df = pd.DataFrame([registro])
        # display(df)
        
        from sqlalchemy import create_engine
        engine = create_engine(
            "mysql+pymysql://root:senha@localhost:3306/RECLAMEAQUI_CONSORCIO"
        )

        # envio
        df.to_sql(
            name="reclamacoes_detalhe",
            con=engine,
            if_exists="append",
            index=False
        )
        print("Dados enviados com sucesso!")
        display(df)

finally:
    driver.quit()
```
