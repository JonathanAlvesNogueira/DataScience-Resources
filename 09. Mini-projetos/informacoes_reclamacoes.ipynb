```python
import time
import random
import undetected_chromedriver as uc

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = uc.Chrome(version_main=148)
wait = WebDriverWait(driver, 15)

dados = []
lista = [
    "https://www.reclameaqui.com.br/consorcio-fiat-e-itau/demora-excessiva-e-falta-de-informacao-no-processo-de-transferencia-de-titularidade-de-consorcio-resultando-em-risco-de-perda-de-negocio_I4xiLCnb0Tc8Uqfx/"
]
try:
    for url in lista:
        print(f"Acessando {url}...")
        driver.get(url)

        wait.until(
            EC.presence_of_all_elements_located((By.TAG_NAME, "p"))
        )

        time.sleep(random.uniform(3, 6))

        elementos = driver.find_elements(By.TAG_NAME, "p")

        for el in elementos:
            texto = el.text.strip()
            if texto and len(texto) > 10:
                dados.append({
                    "texto": texto
                })
finally:
    driver.quit()

display(dados)
```

```python
registro = {
    "ds_situacao_reclamacao": dados[0]["texto"],
    "estado_reclamacao": dados[1]["texto"],
    "data_reclamacao": dados[2]["texto"],
    "id_reclamacao": dados[3]["texto"].replace("ID: ", ""),
    "descricao_reclamacao": dados[4]["texto"]
}

display(registro)
import pandas as pd
df = pd.DataFrame([registro])
display(df)
```
