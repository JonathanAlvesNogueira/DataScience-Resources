```python
import time
import random
import undetected_chromedriver as uc

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException

driver = uc.Chrome(version_main=148)
wait = WebDriverWait(driver, 15)

dados = []

try:
    for pagina in range(2, 10):

        url = f"https://www.reclameaqui.com.br/empresa/consorcio-fiat-e-itau/lista-reclamacoes/?pagina={pagina}"
        print(f"Acessando página {pagina}...")

        driver.get(url)

        wait.until(
            EC.presence_of_all_elements_located((By.TAG_NAME, "a"))
        )

        time.sleep(random.uniform(3, 6))

        elementos = driver.find_elements(By.TAG_NAME, "a")

        for i in range(len(elementos)):
            try:
                # 🔥 rebusca o elemento pelo índice
                elementos = driver.find_elements(By.TAG_NAME, "a")
                texto = elementos[i].text.strip()

                if texto and len(texto) > 40:
                    dados.append({"titulo": texto})

            except StaleElementReferenceException:
                # ignora e segue
                continue

        print(f"Página {pagina} coletada. Total até agora: {len(dados)}")

        time.sleep(random.uniform(5, 10))

finally:
    driver.quit()
print(f"\nTotal coletado: {len(dados)}")
```

### Código pronto

```python
import time
import random
import undetected_chromedriver as uc

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException

driver = uc.Chrome(version_main=148)
wait = WebDriverWait(driver, 15)

dados = []

try:
    for pagina in range(2, 3):

        url = f"https://www.reclameaqui.com.br/empresa/consorcio-fiat-e-itau/lista-reclamacoes/?pagina={pagina}"
        print(f"Acessando página {pagina}...")

        driver.get(url)

        wait.until(
            EC.presence_of_all_elements_located((By.TAG_NAME, "a"))
        )

        time.sleep(random.uniform(3, 6))

        elementos = driver.find_elements(By.TAG_NAME, "a")

        for i in range(len(elementos)):
            try:
                # 🔥 rebusca o elemento pelo índice
                elementos = driver.find_elements(By.TAG_NAME, "a")
                texto = elementos[i].text.strip()
                link = elementos[i].get_attribute("href")

                # spans = driver.find_elements(By.TAG_NAME, "span")
                # texto_span = spans[i].text.strip()
                # if texto_span and "há" in texto_span.lower():


                if texto and len(texto) > 40 and link:
                    dados.append({
                        "titulo": texto,
                        "link": link
                    })

            except StaleElementReferenceException:
                # ignora e segue
                continue

        print(f"Página {pagina} coletada. Total até agora: {len(dados)}")

        time.sleep(random.uniform(5, 10))

finally:
    driver.quit()
print(f"\nTotal coletado: {len(dados)}")
```
