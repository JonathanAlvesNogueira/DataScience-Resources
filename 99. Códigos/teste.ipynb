```python
import pandas as pd 
import numpy as np
```

```python
# Configura o tamanho da base
n = 1_000_000

# Gera dados sintéticos
np.random.seed(42)

df = pd.DataFrame({
    'ClienteID': np.random.randint(1000, 5000, n),
    'Produto': np.random.choice(['Cartão', 'Empréstimo', 'Cheque Especial', 'Financiamento'], n),
    'DataTransacao': pd.to_datetime('2024-01-01') + pd.to_timedelta(np.random.randint(0, 365, n), unit='d'),
    'Valor': np.random.randint(50, 5000, n),
    'ScoreCredito': np.random.randint(300, 850, n),
    'Inadimplente': np.random.choice([0, 1], n, p=[0.9, 0.1])
})

# # Salvar em CSV para treinar
# df.to_csv('base_treinamento_credito.csv', index=False)

display(df.head())
```

```python
def perc_inadimplente(x):
    return (x == 1).sum() / len(x) * 100

tabela = pd.pivot_table(
    df,
    index='Produto',
    values='Inadimplente',
    aggfunc=perc_inadimplente
)

display(tabela)
```

```python
# Função para colorir em escala de cores
def color_scale(s):
    # Normaliza entre 0 e 1
    norm = (s - s.min()) / (s.max() - s.min())
    # Cria cor interpolando entre verde e vermelho
    colors = ['#00FF00', '#FF0000']  # verde → vermelho
    return ['background-color: {}'.format(
                f'rgba({int(255*norm_val)}, {int(255*(1-norm_val))}, 0, 0.6)'
            ) for norm_val in norm]

# Aplica a formatação à tabela
tabela_estilizada = tabela.style.apply(color_scale)
tabela_estilizada
```
